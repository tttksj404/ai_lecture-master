## 학습 개요

1. **학습 주제**
    - **프롬프팅 기법(Prompting Techniques)**: Zero-shot, Few-shot, CoT(Chain-of-Thought) 등 다양한 프롬프팅 기법의 원리와 적용 방법 이해
    - **합성 데이터 생성(Synthetic Data Generation)**: LLM을 활용한 학습 데이터 증강 및 다양성 확보 전략
    - **LLM as a Judge**: 생성된 데이터의 품질을 LLM으로 자동 평가하고 선별하는 파이프라인 구축

2. **학습 목표**
    - Zero-shot, Few-shot, CoT 등 **다양한 프롬프팅 기법**의 차이를 이해하고, 상황에 맞게 적용할 수 있다.
    - LLM API를 활용하여 **Text-to-SQL task를 위한 한국어 학습 데이터**를 생성할 수 있다.

    - LLM as a Judge 방식으로 **생성된 데이터의 품질을 자동 평가**하고 필터링할 수 있다.
    - 프롬프팅 기법에 따른 **생성 품질/다양성/일관성을 비교 분석**할 수 있다.

3. **핵심 개념**
    - **Zero-shot Prompting**: 예시 없이 지시문만으로 모델에게 작업을 요청하는 기법. 모델의 사전 학습된 지식에 의존
    - **Few-shot Prompting**: 소수의 예시(1~5개)를 프롬프트에 포함하여 모델이 패턴을 학습하도록 유도하는 기법
    - **Chain-of-Thought (CoT)**: 복잡한 추론 과제에서 단계별 사고 과정을 명시적으로 유도하여 정확도를 높이는 기법
    - **합성 데이터(Synthetic Data)**: 실제 데이터를 기반으로 LLM을 통해 생성한 인공 데이터로, 학습 데이터 증강에 활용
    - **LLM as a Judge**: LLM을 평가자로 활용하여 생성된 콘텐츠의 품질을 자동으로 판단하는 방법론
    - **Structured Output (JSON Schema)**: API 응답을 정형화된 JSON 형식으로 받아 안정적인 후처리를 가능하게 하는 기법

## Step 1: 프롬프팅 기법 비교 실습

### Concept Check

#### 프롬프팅 기법의 종류와 특징

LLM의 성능은 프롬프트를 어떻게 작성하느냐에 크게 좌우됩니다. 지난 시간에 배운 대표적인 프롬프팅 기법을 복습해보겠습니다.

**1. Zero-shot Prompting**
- 예시 없이 지시문만으로 작업을 요청
- 간단하고 빠르지만, 복잡한 작업에서는 정확도가 낮을 수 있음
```
다음 영어를 한국어로 번역해주세요: "Hello, world!"
```

**2. Few-shot Prompting**
- 1~5개의 예시를 프롬프트에 포함
- 모델이 패턴을 파악하여 더 일관된 결과 생성
```
영어를 한국어로 번역합니다.

English: Good morning
Korean: 좋은 아침이에요

English: Thank you
Korean: 감사합니다

English: Hello, world!
Korean:
```

**3. Chain-of-Thought (CoT) Prompting**
- 단계별 추론 과정을 명시적으로 유도
- 복잡한 추론이 필요한 작업에서 효과적
```
다음 영어를 한국어로 번역해주세요. 번역 전에 문장의 핵심 의미를 먼저 파악하고, 한국어로 자연스럽게 표현해주세요.

1단계: 원문 분석 - 주어, 동사, 목적어 파악
2단계: 핵심 의미 추출
3단계: 한국어로 자연스럽게 변환
```

#### 프롬프팅 기법 선택 가이드

| 상황 | 권장 기법 |
|------|----------|
| 간단한 작업, 빠른 응답 필요 | Zero-shot |
| 특정 형식/스타일 필요 | Few-shot |
| 복잡한 추론, 정확성 중요 | CoT |

먼저 실습에 사용할 데이터셋을 로드합니다.

````python
# Text-to-SQL 데이터셋 로드
dataset_name = "b-mc2/sql-create-context"
dataset = load_dataset(dataset_name, split="train")

# 비용 절감을 위해 10개 샘플만 선택
dataset_10 = dataset.select(range(10))
print(f"데이터셋 크기: {len(dataset_10)}개 샘플")
print(f"\n첫 번째 샘플:")
dataset_10[0]
````

LLM 응답을 JSON 형식으로 받기 위한 스키마를 다음과 같이 정의합니다. 지난 시간에는 System Prompt에 직접 출력 포맷을 문자열로 정의해줬다면, 이번에는 `.create(response_format=response_format)`을 통해 좀 더 명확하게 출력 포맷을 지정해줍니다. 이렇게 수행하면 다음과 같은 이점이 있습니다.
- 단순 권장사항을 넘어 변동 없는 응답 포맷을 받을 수 있습니다.
- 추가 처리의 필요성이 줄어 파싱 에러 같은 오류가 줄어듭니다.

````python
# 번역 결과를 JSON으로 받기 위한 response_format 정의
translation_response_format = {
    "type": "json_schema",
    "json_schema": {
        "name": "translation",
        "strict": True,
        "schema": {
            "type": "object",
            "properties": {
                "korean": {
                    "type": "string",
                    "description": "한국어로 번역된 문장"
                }
            },
            "required": ["korean"],
            "additionalProperties": False
        }
    }
}
````

### TODO 1: Zero-shot 프롬프팅으로 번역 수행

**요구사항:**
1. 예시 없이 지시문만으로 영어를 한국어로 번역하는 시스템 프롬프트(`ZERO_SHOT_PROMPT`)를 작성하고, 출력 결과를 `zero_shot_result`에 저장해주세요.
2. 첫 번째 샘플 `test_question`을 번역하고 결과를 출력하세요.
   1. 이 때 위에서 제작한 `translation_response_format`을 활용하여 출력 포맷을 지정해주세요.

````python
# [문제였던 부분 반영]
# 1. 예시 없이 지시문만으로 영어를 한국어로 번역하는 시스템 프롬프트(`ZERO_SHOT_PROMPT`)를 작성하고
# 출력 결과를 `zero_shot_result`에 저장해주세요.
ZERO_SHOT_PROMPT = """다음 영어 문장을 한국어로 번역해주세요.
원문의 의미를 정확히 전달하면서 자연스러운 한국어로 번역하세요."""

# 2. 첫 번째 샘플 `test_question`을 번역하고 결과를 출력하세요.
test_question = dataset_10[0]["question"]
print(f"원문: {test_question}")

response = client.chat.completions.create(
    model="solar-pro3",
    messages=[
        {"role": "system", "content": ZERO_SHOT_PROMPT},
        {"role": "user", "content": test_question}
    ],
    response_format=translation_response_format
)

zero_shot_result = json.loads(response.choices[0].message.content)
print(f"Zero-shot 번역: {zero_shot_result['korean']}")
````

### TODO 2: Few-shot 프롬프팅으로 번역 품질 개선

**요구사항**: 2~3개의 번역 예시를 포함한 시스템 프롬프트(`FEW_SHOT_PROMPT`)를 작성하고 출력 결과를 `few_shot_result`에 저장해주세요.

````python
# [문제였던 부분 반영]
# 2~3개의 번역 예시를 포함한 시스템 프롬프트(`FEW_SHOT_PROMPT`)를 작성하고
# 출력 결과를 `few_shot_result`에 저장해주세요.
FEW_SHOT_PROMPT = """다음 영어 문장을 한국어로 번역해주세요. 아래 예시를 참고하여 자연스러운 한국어로 번역하세요.

예시 1:
English: What is the total number of employees?
Korean: 직원의 총 수는 몇 명인가요?

예시 2:
English: Show the names of customers who made purchases in January.
Korean: 1월에 구매한 고객들의 이름을 보여주세요.

예시 3:
English: How many products have a price greater than 100?
Korean: 가격이 100보다 큰 상품은 몇 개인가요?

위 예시와 같은 스타일로 번역해주세요."""

response = client.chat.completions.create(
    model="solar-pro3",
    messages=[
        {"role": "system", "content": FEW_SHOT_PROMPT},
        {"role": "user", "content": test_question}
    ],
    response_format=translation_response_format
)
few_shot_result = json.loads(response.choices[0].message.content)

# 결과 확인 및 비교
print(f"원문: {test_question}")
print(f"Few-shot 번역: {few_shot_result['korean']}")
print(f"\n[비교] Zero-shot: {zero_shot_result['korean']}")
````

````python
# [문제였던 부분 반영]
# 단계별 추론 과정을 유도하는 시스템 프롬프트(`COT_PROMPT`)를 작성하고
# 출력 결과를 `cot_result`에 저장해주세요.
COT_PROMPT = """다음 영어 문장을 한국어로 번역해주세요. 번역 시 다음 단계를 따라 진행하세요:

1단계 - 문장 구조 분석: 주어, 동사, 목적어 등 문장의 핵심 요소를 파악합니다.
2단계 - 핵심 의미 추출: 문장이 전달하고자 하는 핵심 의미를 파악합니다.
3단계 - 한국어 표현 선택: 한국어에서 자연스럽게 사용되는 표현을 선택합니다.
4단계 - 최종 번역: 위 분석을 바탕으로 자연스러운 한국어 문장을 완성합니다.

최종 번역 결과만 출력해주세요."""

response = client.chat.completions.create(
    model="solar-pro3",
    messages=[
        {"role": "system", "content": COT_PROMPT},
        {"role": "user", "content": test_question}
    ],
    response_format=translation_response_format
)

cot_result = json.loads(response.choices[0].message.content)

# 결과 확인 및 비교
print(f"원문: {test_question}")
print(f"CoT 번역: {cot_result['korean']}")
print(f"\n=== 프롬프팅 기법별 결과 비교 ===")
print(f"Zero-shot: {zero_shot_result['korean']}")
print(f"Few-shot:  {few_shot_result['korean']}")
print(f"CoT:       {cot_result['korean']}")
````

## Step 2: 데이터 증강을 위한 Prompt Engineering

### Concept Check

#### 왜 데이터 증강이 필요한가?

머신러닝 모델을 학습시킬 때 **다양하고 충분한 양의 데이터**가 필요합니다. 하지만 현실에서는:

1. **한국어 데이터 부족**: 영어 대비 공개된 한국어 데이터셋이 매우 적음
2. **데이터 수집 비용**: 고품질 라벨링 데이터 확보에 많은 시간과 비용 필요
3. **다양성 부족**: 실제 사용자의 다양한 표현 방식을 커버하지 못함

**데이터 증강(Data Augmentation)**은 기존 데이터를 변형하거나 합성하여 학습 데이터의 양과 다양성을 늘리는 기법입니다.

#### 페르소나 기반 데이터 증강

동일한 의미의 문장도 화자에 따라 다르게 표현됩니다:
- 초등학생: "56살보다 많은 부서장 몇 명이에요?"
- 직장인: "56세 이상인 부서장 수는 몇 명입니까?"
- 친구 사이: "부서장들 중 56살 넘은 사람 몇 명이나 되지?"

이렇게 **페르소나(Persona)**를 활용하면 동일한 원본 데이터에서 다양한 스타일의 학습 데이터를 생성할 수 있습니다.

#### 데이터 증강 파이프라인

효율적인 데이터 증강을 위해 다음 순서를 따릅니다:

1. **소규모 테스트** (10개): 프롬프트와 코드 검증
2. **중규모 확장** (100개): 품질 일관성 확인
3. **대규모 생성** (1000개+): 전체 파이프라인 실행

이 방식으로 **비용을 최소화**하면서 **품질을 보장**할 수 있습니다.

### TODO 4: 페르소나 기반 다양한 스타일 번역

**요구사항:**
1. 5개 이상의 다양한 페르소나를 `personas` 리스트에 정의하세요.
   1. 페르소나는 영어로 작성해도 됩니다. 다양한 연령대, 직업, 말투를 고려하세요.
2. 페르소나를 반영한 시스템 프롬프트 템플릿(`PERSONA_PROMPT_TEMPLATE`)을 작성하세요.
3. 모든 페르소나로 첫 번째 샘플에 대해 번역을 수행해 `translated_results`에 할당해주세요.

````python
# [문제였던 부분 반영]
# 1. 5개 이상의 다양한 페르소나를 `personas` 리스트에 정의하세요
personas = [
    "Friendly Adult - Kind and polite tone, like an older sibling giving advice.",
    "Elementary Student - Simple, cheerful, and naive tone, like a 10-year-old child.",
    "Formal Business - Polite, professional tone suitable for documents or work emails.",
    "Casual Friend - Relaxed, informal tone, as if chatting with a close friend.",
    "Elderly Person - Gentle, old-fashioned, and wise tone, like a kind grandparent.",
]

# 2. 페르소나를 반영한 시스템 프롬프트 템플릿(`PERSONA_PROMPT_TEMPLATE`)을 작성하세요
PERSONA_PROMPT_TEMPLATE = """You are a translation assistant.
Your task is to translate English sentences into Korean.
You must keep the meaning accurate, but adapt the tone and style according to the requested persona.

Persona: {persona}

Translate the given English sentence into Korean with the specified persona's tone and style.
"""

# 3. 모든 페르소나로 첫 번째 샘플에 대해 번역을 수행해 `translated_results`에 할당해주세요.
translated_results = []
test_row = dataset_10[0]

print(f"원문: {test_row['question']}\n")
print("=" * 60)

for persona in personas:
    system_prompt = PERSONA_PROMPT_TEMPLATE.format(persona=persona)

    response = client.chat.completions.create(
        model="solar-pro3",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": test_row["question"]}
        ],
        response_format=translation_response_format
    )

    result = json.loads(response.choices[0].message.content)
    translated_results.append({
        "persona": persona,
        "korean": result["korean"]
    })

    persona_short = persona.split(" - ")[0]
    print(f"[{persona_short}]")
    print(f"  {result['korean']}\n")

print(f"\n총 {len(translated_results)}개의 다양한 번역 생성 완료!")
````

여러분이 정의한 프롬프트를 토대로 전체 데이터셋에 대해 페르소나 기반 증강을 수행합니다. (데모용으로 첫 번째 샘플만 실행)

````python
# 전체 데이터셋 증강 (데모: 1개 샘플만)
augmented_dataset = []

for idx, row in enumerate(dataset_10):
    sample_translations = {
        "id": idx,
        "original_question": row["question"],
        "context": row["context"],
        "answer": row["answer"],
        "translations": []
    }

    for persona in personas:
        system_prompt = PERSONA_PROMPT_TEMPLATE.format(persona=persona)

        response = client.chat.completions.create(
            model="solar-pro3",
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": row["question"]}
            ],
            response_format=translation_response_format
        )

        result = json.loads(response.choices[0].message.content)
        sample_translations["translations"].append({
            "persona": persona,
            "korean": result["korean"]
        })

    augmented_dataset.append(sample_translations)
    print(f"샘플 {idx + 1}/{len(dataset_10)} 증강 완료")

    # 데모를 위해 1개만 실행 (전체 실행 시 이 줄 제거)
    break

print(f"\n증강된 데이터셋 구조:")
print(json.dumps(augmented_dataset[0], indent=2, ensure_ascii=False))
````

## Step 3: 증강 데이터 평가 (LLM as a Judge)

### Concept Check

#### 왜 생성 데이터를 평가해야 하는가?

LLM으로 생성한 데이터가 항상 고품질은 아닙니다. 발생할 수 있는 문제들:

1. **의미 왜곡**: 번역 과정에서 원래 의미가 변질될 수 있음
2. **환각(Hallucination)**: 원문에 없는 정보가 추가될 수 있음
3. **품질 불일치**: 일부 결과물의 품질이 현저히 낮을 수 있음

따라서 **생성된 데이터의 품질을 평가하고 필터링**하는 과정이 필수적입니다.

#### LLM as a Judge

**LLM을 평가자(Judge)로 활용**하는 방법론입니다:

```
[평가 프롬프트]
원문: How many heads of the departments are older than 56?
번역: 56세 이상인 부서장은 몇 명인가요?

위 번역이 원문의 의미를 정확히 전달하는지 0-10점으로 평가하고, 이유를 설명해주세요.
```

**장점:**
- 대량의 데이터를 자동으로 평가 가능
- 사람 평가 대비 비용/시간 절약
- 일관된 기준으로 평가 가능

**주의점:**
- LLM 평가도 완벽하지 않음 (일부 샘플은 사람이 검증 필요)
- 평가 프롬프트의 품질이 결과에 큰 영향

### TODO 5: LLM as a Judge 프롬프트 작성

**요구사항:**
1. 원본 영어와 번역된 한국어를 비교하여 의미 보존 여부를 평가하는 시스템 프롬프트(`JUDGE_SYSTEM_PROMPT`)를 작성하세요
2. 평가 결과를 JSON으로 받기 위한 `evaluate_response_format`을 정의하세요. 다음 내용을 포함해주세요.
   - `score`: 0-10 사이의 정수 점수
   - `reason`: 평가 이유 설명
3. 평가용 사용자 프롬프트 템플릿 `JUDGE_USER_TEMPLATE` 를 작성해주세요.

````python
# [문제였던 부분 반영]
# 1. 원본 영어와 번역된 한국어를 비교하여
# 의미 보존 여부를 평가하는 시스템 프롬프트(`JUDGE_SYSTEM_PROMPT`)를 작성하세요
JUDGE_SYSTEM_PROMPT = """You are an evaluation assistant.
Your task is to judge whether a Korean translation preserves the original English meaning, while only changing the tone/persona (e.g., friendly, childish, formal, humorous).

Evaluation Criteria:
1. Meaning Preservation (0-10 points)
   - Does the Korean text accurately convey the original English meaning?
   - No key information should be added, removed, or distorted.
   - 10: Perfect meaning preservation
   - 7-9: Minor differences but core meaning intact
   - 4-6: Some meaning lost or altered
   - 0-3: Significant meaning distortion

Instructions:
- Always compare the English source with the Korean translation.
- Do NOT allow stylistic creativity to change the factual content.
- If the meaning is preserved and only the tone changes, give a high score.
- Provide a brief reason for your score in Korean.
"""

# 2. 평가 결과를 JSON으로 받기 위한 `evaluate_response_format`을 정의하세요. 다음 내용을 포함해주세요.
evaluate_response_format = {
    "type": "json_schema",
    "json_schema": {
        "name": "evaluation",
        "strict": True,
        "schema": {
            "type": "object",
            "properties": {
                "score": {
                    "type": "integer",
                    "description": "Evaluation score from 0 to 10"
                },
                "reason": {
                    "type": "string",
                    "description": "Brief explanation for the score in Korean"
                }
            },
            "required": ["score", "reason"],
            "additionalProperties": False
        }
    }
}

# 3. 평가용 사용자 프롬프트 템플릿 `JUDGE_USER_TEMPLATE` 를 작성해주세요.
JUDGE_USER_TEMPLATE = """Please evaluate the following translation:

- Original (English): {english}
- Translation (Korean): {korean}
"""

# 단일 번역 평가 테스트
test_english = test_row["question"]
test_korean = translated_results[0]["korean"]  # 첫 번째 페르소나 번역

response = client.chat.completions.create(
    model="solar-pro3",
    messages=[
        {"role": "system", "content": JUDGE_SYSTEM_PROMPT},
        {"role": "user", "content": JUDGE_USER_TEMPLATE.format(
            english=test_english,
            korean=test_korean
        )}
    ],
    response_format=evaluate_response_format
)

eval_result = json.loads(response.choices[0].message.content)
print(f"원문: {test_english}")
print(f"번역: {test_korean}")
print(f"\n평가 점수: {eval_result['score']}/10")
print(f"평가 이유: {eval_result['reason']}")
````

### TODO 6: 평가 점수 기반 데이터 필터링

**요구사항:**
1. 모든 번역 결과에 대해 평가를 수행하여 `evaluated_translations`에 저장해주세요.
2. 여러분이 지정한 `filtering_score` 이상의 점수를 받은 데이터만 필터링하여 `filtered_dataset`에 저장해주세요.

````python
# [문제였던 부분 반영]
# 1. 모든 번역 결과에 대해 평가를 수행하여 `evaluated_translations`에 저장해주세요.
evaluated_translations = []

print(f"원문: {test_row['question']}\n")
print("=" * 60)

for translation in translated_results:
    response = client.chat.completions.create(
        model="solar-pro3",
        messages=[
            {"role": "system", "content": JUDGE_SYSTEM_PROMPT},
            {"role": "user", "content": JUDGE_USER_TEMPLATE.format(
                english=test_row["question"],
                korean=translation["korean"]
            )}
        ],
        response_format=evaluate_response_format
    )

    eval_result = json.loads(response.choices[0].message.content)

    evaluated_item = {
        "persona": translation["persona"],
        "korean": translation["korean"],
        "score": eval_result["score"],
        "reason": eval_result["reason"]
    }
    evaluated_translations.append(evaluated_item)

    persona_short = translation["persona"].split(" – ")[0]
    print(f"[{persona_short}] 점수: {eval_result['score']}/10")
    print(f"  번역: {translation['korean']}")
    print(f"  이유: {eval_result['reason']}\n")


# 2. 여러분이 지정한 `filtering_score` 이상의 점수를 받은 데이터만 필터링하여
# `filtered_dataset`에 저장해주세요.
filtering_score = 8  # 8점 이상만 통과
filtered_dataset = []

for item in evaluated_translations:
    if item["score"] >= filtering_score:
        filtered_item = {
            "question": item["korean"],
            "context": test_row["context"],
            "answer": test_row["answer"],
            "score": item["score"]
        }
        filtered_dataset.append(filtered_item)

print(f"\n=== 필터링 결과 ===")
print(f"필터링 기준: {filtering_score}점 이상")
print(f"원본 데이터: {len(evaluated_translations)}개")
print(f"필터링 후: {len(filtered_dataset)}개")
print(f"통과율: {len(filtered_dataset)/len(evaluated_translations)*100:.1f}%")

print(f"\n=== 최종 합성 데이터셋 ===")
for i, item in enumerate(filtered_dataset):
    print(f"\n[{i+1}] 점수: {item['score']}/10")
    print(f"    질문: {item['question']}")
    print(f"    답변: {item['answer']}")
````
