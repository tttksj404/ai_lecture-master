## 학습 개요

### 학습 주제
- **Supervisor 패턴 이해**: 관리자 Agent가 여러 Worker Agent를 조율하는 협업 방식 학습
- **관리자-작업자 구조 설계**: 작업 분배와 결과 통합을 담당하는 Supervisor 역할 이해
- **실습 결과와 비교 분석**: Planner-Worker 패턴과 Supervisor 패턴의 차이점 분석

### 학습 목표
- Supervisor 패턴을 이해하고, 실습에서 배운 패턴과 비교하여 상황에 적합한 패턴을 선택할 수 있다
  - Supervisor 패턴의 구조와 특징을 설명할 수 있다
  - Supervisor가 여러 Worker를 조율하는 구조를 구현할 수 있다
  - 실습(Planner-Worker)과 과제(Supervisor) 패턴의 차이점을 분석할 수 있다
  - LangGraph StateGraph를 사용하여 다중 Worker 워크플로우를 구성할 수 있다
  - 상황에 따라 적합한 Multi-Agent 패턴을 선택하고 그 이유를 설명할 수 있다

### 핵심 개념
- **Supervisor 패턴**: 관리자 Agent가 여러 Worker Agent를 조율하는 협업 방식. 작업 분배와 결과 통합을 Supervisor가 담당하며, 각 Worker는 전문 영역에 집중한다. 병렬 처리가 필요한 독립 작업에 적합하다.
- **Worker 분배**: Supervisor가 사용자 요청을 분석하여 적절한 Worker에게 작업을 할당하는 과정. 병렬 처리 또는 순차 처리 가능. 작업의 독립성에 따라 분배 전략이 달라진다.
- **결과 통합**: 여러 Worker의 결과를 Supervisor가 취합하여 일관성 있는 최종 응답을 생성하는 과정. 각 결과 간의 연결성과 정합성을 확보하는 것이 핵심이다.
- **도메인별 전문화**: 각 Worker가 특정 도메인(일정, 예산 등)에 특화된 프롬프트와 역할을 가지는 구조. 전문화를 통해 각 영역의 결과 품질이 향상된다.
- **StateGraph 확장**: 여러 노드를 연결하고 상태를 공유하는 LangGraph 워크플로우 구성 방법. 다중 Worker 구조에서는 State 필드를 Worker별로 구분하여 관리한다.

# Step 2: Supervisor 패턴 이해

### Concept Check: Supervisor 패턴이란?

**왜 이 단계가 필요한가요?**  
Supervisor 패턴은 Planner-Worker와 다른 협업 방식입니다. 여러 Worker가 **도메인별로 전문화**되어 있고, Supervisor가 **작업 분배와 결과 통합**을 담당합니다. 이 구조를 이해해야 적절히 구현할 수 있습니다.

**Supervisor 패턴**은 관리자 Agent가 여러 Worker Agent를 조율하는 협업 방식입니다.

```
사용자 요청
    ↓
[Supervisor Agent] → 작업 분배 (Fan-out)
    ↓
┌───────────┬───────────┐
│  Worker A │  Worker B │  ← 병렬 실행
│ (일정담당) │ (예산담당) │
└───────────┴───────────┘
    ↓
[Supervisor 통합] → 결과 통합 (Fan-in: 모든 Worker 완료 대기)
    ↓
최종 결과
```

#### Supervisor Agent의 역할
- 사용자 요청을 분석하여 작업 분배
- 각 Worker에게 적절한 작업 할당 (병렬 실행)
- Worker들의 결과를 취합하여 최종 응답 생성

#### 실습(Planner-Worker + Reflection) vs 과제(Supervisor) 비교

| 항목 | Planner-Worker + Reflection | Supervisor |
|------|------------------------------|------------|
| 구조 | 순차적 (계획→실행→검토) | **병렬적** (분배→실행→통합) |
| 역할 분리 | 기능별 (계획/실행/검토) | 도메인별 (일정/예산 등) |
| 장점 | 품질 보장, 단계별 검증 | **병렬 처리**, 전문화 가능 |
| 적합한 상황 | 순차적 작업 흐름 | **독립적 하위 작업**이 있는 경우 |

#### Agent 간 메시지 흐름 예시 (병렬 실행)
```
[User] "서울 2박3일 여행 계획"
    ↓
┌───────────────────────────────────────┐
│  (병렬 실행)                           │
│  [Schedule Worker]    [Budget Worker] │
│  State 읽기:          State 읽기:      │
│  user_request         user_request    │
│       ↓                    ↓          │
│  LLM 호출              LLM 호출        │
│       ↓                    ↓          │
│  State 쓰기:          State 쓰기:      │
│  schedule_result      budget_result   │
└───────────────────────────────────────┘
    ↓ (Fan-in: 두 Worker 모두 완료 대기)
[Supervisor 통합] State 읽기: schedule_result, budget_result
    ↓ LLM 호출 (통합)
[Supervisor 통합] State 쓰기: final_result = "## 최종 여행 계획\n..."
```

### Guided Build: Supervisor 패턴 구조 다이어그램

| 구성 요소 | 역할 |
|----------|------|
| Supervisor 통합 | 작업 분배, 결과 통합, 최종 응답 생성 |
| Worker A (일정) | 여행 일정 및 관광지 담당 |
| Worker B (예산) | 예산 계산 및 맛집 추천 담당 |

#### 병렬 실행 워크플로우 구조

```
        ┌─→ [Schedule Worker] ─┐
START ──┤       (병렬 실행)      ├──→ [Supervisor 통합] → END
        └─→ [Budget Worker] ───┘
```

- **Fan-out**: START에서 두 Worker로 동시에 작업 분배
- **Fan-in**: 두 Worker가 모두 완료되면 Supervisor가 결과 통합

### TODO 1: SupervisorState 정의

- **요구사항**: Supervisor 패턴에 필요한 State를 정의하세요.
- **입력**: 없음
- **출력**: 
  - `SupervisorState` (TypedDict): State 스키마 클래스
    - `user_request` (str): 사용자 요청
    - `schedule_result` (str): 일정 Worker 결과
    - `budget_result` (str): 예산 Worker 결과
    - `final_result` (str): Supervisor 통합 결과
    - `messages` (Annotated[List[dict], operator.add]): 메시지 로그 (병렬 실행 시 자동 병합)
- **예상 결과**: TypedDict 클래스가 정의되어 StateGraph에서 사용 가능
- **확인 방법**: `print(SupervisorState.__annotations__)`으로 필드 확인
- **힌트**: 
  - 병렬 실행 시 State 충돌을 방지하려면 `Annotated`와 `operator.add` 사용
  - `messages: Annotated[List[dict], operator.add]`로 정의하면 병렬 Worker의 메시지가 자동 병합
  - 필요한 import 구문:
    ```python
    from typing import TypedDict, List, Annotated
    import operator  # operator.add 사용을 위해 필요
    ```

````python
# [문제였던 부분 반영]
# 스켈레톤: 기본 Agent 클래스 + LangGraph State
from typing import Dict, Any, List, Annotated
from typing import TypedDict
import operator

# LangGraph State 정의 (병렬 실행을 위한 Annotated reducer 사용)
class SupervisorState(TypedDict):
    """Supervisor 패턴의 상태 (병렬 실행 지원)"""
    user_request: str
    schedule_result: str    # 일정 Worker 결과
    budget_result: str      # 예산 Worker 결과
    final_result: str       # Supervisor가 통합한 최종 결과
    # 병렬 실행 시 messages 리스트를 자동 병합하기 위해 Annotated + operator.add 사용
    messages: Annotated[List[dict], operator.add]

print("SupervisorState 정의 완료 (병렬 실행 지원)")
````

### TODO 2: Schedule Worker 노드 구현

- **요구사항**: 일정/관광지를 담당하는 Worker 노드를 구현합니다.
- **입력**: 
  - `state` (SupervisorState): 현재 상태 딕셔너리
    - `state["user_request"]` (str): 사용자 요청
- **출력**: 
  - `dict`: **변경된 필드만** 포함한 딕셔너리 (병렬 실행 필수!)
    - `schedule_result` (str): 날짜별 일정, 관광지 추천 결과
    - `messages` (List[dict]): 새 메시지 항목만 (리스트로 감싸서 반환)
- **예상 결과**: 날짜별 상세 일정과 관광지 정보 (입장료, 소요시간 포함)
- **확인 방법**: `print(result.get("schedule_result", "미구현"))`
- **핵심**: 병렬 실행을 위해 `{**state, ...}` 대신 변경 필드만 반환

> **⚠️ 병렬 실행 시 State 덮어쓰기 위험**
> 
> 병렬로 실행되는 Worker가 `{**state, "schedule_result": ...}` 처럼 전체 State를 복사해서 반환하면, 다른 Worker가 작성한 결과를 덮어쓸 수 있습니다. 예를 들어 Schedule Worker가 전체 State를 반환하면서 `budget_result: ""`를 포함하면, Budget Worker가 이미 작성한 결과가 사라집니다.
> 
> **해결책**: 각 Worker는 자신이 변경한 필드만 반환합니다. LangGraph는 반환된 필드만 State에 병합(merge)하므로, 다른 Worker의 결과는 보존됩니다.

````python
# [문제였던 부분 반영]
# TODO: Schedule Worker 노드 구현

def schedule_worker_node(state: SupervisorState) -> dict:
    """
    Schedule Worker: 여행 일정 및 관광지 담당
    (병렬 실행을 위해 변경된 필드만 반환)
    """
    user_request = state["user_request"]
    
    prompt = f"""
당신은 여행 일정을 담당하는 전문 Agent입니다.

다음 요청에서 여행 일정과 관광지 관련 부분만 처리하세요:
- 날짜별 상세 일정
- 관광지 추천 (입장료, 소요시간 포함)
- 이동 동선 고려

요청: {user_request}

여행 일정과 관광지 정보만 작성하세요. 예산이나 맛집은 다른 Agent가 담당합니다.
"""
    
    response = llm.invoke(prompt)
    
    # 병렬 실행: 변경된 필드만 반환 (messages는 Annotated로 자동 병합)
    return {
        "schedule_result": response.content,
        "messages": [{"agent": "Schedule Worker", "output": response.content, "type": "schedule"}]
    }

print("Schedule Worker Node 정의 완료")
````

### TODO 4: Supervisor 통합 노드 구현

- **요구사항**: 두 Worker의 결과를 통합하는 Supervisor 통합 노드를 구현합니다.
- **입력**: 
  - `state` (SupervisorState): 현재 상태 딕셔너리
    - `state["schedule_result"]` (str): 일정 Worker 결과
    - `state["budget_result"]` (str): 예산 Worker 결과
    - `state["user_request"]` (str): 원본 사용자 요청
- **출력**: 
  - `dict`: 변경된 필드만 포함한 딕셔너리
    - `final_result` (str): 두 결과를 통합한 최종 여행 계획
    - `messages` (List[dict]): 새 메시지 항목만 (리스트로 감싸서 반환)
- **예상 결과**: 날짜별 일정+맛집+예산이 연결된 완성도 높은 여행 계획
- **확인 방법**: `print(result.get("final_result", "미구현"))`
- **핵심**: 두 Worker의 결과를 자연스럽게 통합 (동선과 예산 연결)

````python
# [문제였던 부분 반영]
# TODO: Supervisor 통합 노드 구현

def supervisor_integration_node(state: SupervisorState) -> dict:
    """
    Supervisor 통합: 두 Worker의 결과를 통합하여 최종 응답 생성
    (변경된 필드만 반환)
    """
    user_request = state["user_request"]
    schedule_result = state["schedule_result"]
    budget_result = state["budget_result"]
    
    prompt = f"""
당신은 Supervisor Agent입니다. 두 Worker Agent의 결과를 통합하여 최종 응답을 생성하세요.

원본 요청: {user_request}

---
[일정 Worker 결과]
{schedule_result}

---
[예산 Worker 결과]
{budget_result}

---
위 두 결과를 통합하여 완성도 높은 최종 여행 계획을 작성하세요.
- 날짜별로 일정과 예산을 연결
- 맛집과 관광지를 동선에 맞게 배치
- 전체 예산 요약 포함
"""
    
    response = llm.invoke(prompt)
    
    # 변경된 필드만 반환 (messages는 Annotated로 자동 병합)
    return {
        "final_result": response.content,
        "messages": [{"agent": "Supervisor", "output": response.content, "type": "integration"}]
    }

print("Supervisor 통합 노드 정의 완료")
````

### TODO 5: LangGraph 워크플로우 구성

- **요구사항**: 위에서 구현한 노드들을 연결하여 병렬 실행 워크플로우를 완성하세요.
- **입력**: 
  - 노드 함수들: `schedule_worker_node`, `budget_worker_node`, `supervisor_integration_node`
- **출력**: 
  - 컴파일된 워크플로우 (app)
- **예상 결과**: `START → [schedule_worker, budget_worker] (병렬) → Supervisor 통합 → END` 구조의 워크플로우
- **확인 방법**: `print("워크플로우 생성 완료" if workflow else "미구현")`
- **힌트**: 
  - `from langgraph.graph import START`로 START 노드 import
  - `add_edge(START, "node_name")`으로 Fan-out 구현 (병렬 분기)
  - 두 Worker에서 같은 노드로 엣지를 연결하면 Fan-in 구현 (두 Worker 모두 완료 대기)

````python
# [문제였던 부분 반영]
# TODO: LangGraph 워크플로우 구성
from langgraph.graph import StateGraph, START, END

def create_supervisor_workflow():
    """
    Supervisor 패턴 워크플로우를 생성합니다.
    
    병렬 실행 구조 (Fan-out / Fan-in):
    
            ┌─→ schedule_worker ─┐
    START ──┤                    ├──→ supervisor_integration → END
            └─→ budget_worker ───┘
    
    Returns:
        컴파일된 워크플로우
    """
    workflow = StateGraph(SupervisorState)

    # 노드 추가
    workflow.add_node("schedule_worker", schedule_worker_node)
    workflow.add_node("budget_worker", budget_worker_node)
    workflow.add_node("supervisor_integration", supervisor_integration_node)

    # 병렬 실행 (Fan-out): START에서 두 Worker로 동시에 분기
    workflow.add_edge(START, "schedule_worker")
    workflow.add_edge(START, "budget_worker")

    # Fan-in: 두 Worker가 모두 완료되면 supervisor_integration으로
    # LangGraph는 같은 노드로 들어오는 모든 엣지가 완료될 때까지 대기
    workflow.add_edge("schedule_worker", "supervisor_integration")
    workflow.add_edge("budget_worker", "supervisor_integration")
    workflow.add_edge("supervisor_integration", END)

    # 컴파일
    return workflow.compile()

print("Supervisor 패턴 워크플로우 구성 완료!")
print("흐름: START → [schedule_worker, budget_worker] (병렬) → Supervisor 통합 → END")
````

# Step 5: 패턴 비교 분석

### Concept Check: Supervisor 패턴 vs Planner-Worker + Reflection 패턴

**왜 이 단계가 필요한가요?**  
실습에서 구현한 Planner-Worker + Reflection 패턴과 과제에서 구현한 Supervisor 패턴을 비교하면, 각 패턴의 장단점과 적합한 상황을 깊이 이해할 수 있습니다. 이 비교 분석 능력이 실제 프로젝트에서 패턴을 선택하는 핵심 역량입니다.

#### 패턴 비교표

| 비교 항목 | Planner-Worker + Reflection (실습) | Supervisor (과제) |
|----------|-----------------------------------|-------------------|
| **실행 방식** | 순차적 (계획→실행→검토) | 병렬적 (분배→동시실행→통합) |
| **Worker 수** | 1개 (범용) | 2개 이상 (전문화) |
| **품질 보장** | Reflection 루프로 자동 개선 | Supervisor가 통합 시 품질 조정 |
| **처리 속도** | 느림 (순차 + 반복 가능) | 빠름 (병렬 실행) |
| **적합한 상황** | 높은 품질이 필요한 단일 흐름 작업 | 독립적 하위 작업의 병렬 처리 |
| **State 관리** | 단일 흐름이라 단순 | 병렬 실행이라 충돌 방지 필요 |

#### 핵심 차이: Reflection의 역할

**실습의 Reflection**: Worker 결과를 평가하고, 부족하면 Worker에게 재작업을 요청하는 **품질 개선 루프**
```
[Planner] → [Worker] → [Reflection] ─── 품질 충분? ──→ [END]
                ↑                              │
                └──── 개선 필요 (재실행) ────────┘
```

**과제의 Supervisor**: 여러 Worker 결과를 **한 번에 통합**하여 최종 결과 생성 (반복 없음)
```
START → [Worker A] ─┐
       [Worker B] ──┤→ [Supervisor 통합] → END
```

- 어떤 상황에서 Planner-Worker가, 어떤 상황에서 Supervisor가 더 적합할까요?
