### 핵심 개념
- **도구 변형**: 기존 도구(쿠폰 발급)의 검증 로직 구조를 유지하면서, 새로운 비즈니스 요구사항(환불 처리)에 맞게 조건과 반환값을 수정하는 방법.
- **Agent 평가**: Agent가 올바른 순서로 도구를 호출했는지(도구 순서), 정책을 위반하지 않았는지(정책 준수), 최종 목표를 달성했는지(작업 완료)를 정량적으로 측정하는 지표.
- **CoT StateGraph**: 도구 호출 전에 명시적 추론 단계(Reasoning)를 추가하여 Agent가 체계적으로 문제를 분석하고 계획을 수립하도록 하는 패턴.

````python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings

# ═══════════════════════════════════════════════════════════════
# 모델 초기화 — OpenAI / Solar 중 택 1 (주석 해제하여 전환)
# ═══════════════════════════════════════════════════════════════

# ── Option A: OpenAI (기본) ──
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
embeddings = OpenAIEmbeddings()

# ── Option B: Solar 대안 (주석 해제하여 사용) ──
# from langchain_upstage import ChatUpstage, UpstageEmbeddings
# llm = ChatUpstage(model="solar-pro3")
# embeddings = UpstageEmbeddings(model="embedding-query")
````

````python
# =============================================================================
# 실습에서 구현한 코드 (과제 실행을 위한 준비)
# =============================================================================

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_core.tools import tool
from langgraph.prebuilt import create_react_agent


# 정책 문서 생성
reward_policy = """\n# 보상 정책 (Reward Policy)\n\n## 쿠폰 발급 조건\n1. 배송 지연 시: 최대 10,000원 쿠폰 발급 가능\n2. 상품 불량 시: 최대 20,000원 쿠폰 발급 가능\n3. 단순 변심: 쿠폰 발급 불가\n\n## 쿠폰 발급 한도\n- 1회 최대 발급 금액: 20,000원\n- 월간 누적 한도: 50,000원\n\n## 주의사항\n- 배송 완료 후 7일 이내에만 보상 가능\n- 이미 쿠폰을 받은 주문에는 추가 발급 불가\n"""

shipping_policy = """\n# 배송 정책 (Shipping Policy)\n\n## 배송 상태\n- 주문 접수: 주문이 접수된 상태\n- 배송 준비중: 상품 포장 중\n- 배송 중: 택배사에 인계됨\n- 배송 완료: 고객에게 전달 완료\n- 배송 지연: 예상 배송일을 초과함\n\n## 배송 지연 기준\n- 일반 배송: 결제 후 3영업일 초과 시 지연\n- 새벽 배송: 주문 익일 미도착 시 지연\n"""

# 과제용 환불 정책 추가
refund_policy = """\n# 환불 정책 (Refund Policy)\n\n## 즉시 환불 가능 조건\n1. 배송 전: 100% 환불 가능\n2. 배송 중: 수령 거부 후 100% 환불 가능\n\n## 배송 완료 후 환불\n3. 배송 완료 후 7일 이내: 상품 불량, 오배송 등 판매자 귀책 사유인 경우만 100% 환불 가능\n4. 배송 완료 후 7일 초과: 환불 불가\n\n## 환불 불가 사유\n- 단순 변심 (배송 완료 후)\n- 사용 흔적이 있는 상품\n- 포장이 훼손된 상품\n"""

# 파일로 저장
with open("reward_policy.txt", "w", encoding="utf-8") as f:
    f.write(reward_policy)

with open("shipping_policy.txt", "w", encoding="utf-8") as f:
    f.write(shipping_policy)

with open("refund_policy.txt", "w", encoding="utf-8") as f:
    f.write(refund_policy)

# 문서 로드
loaders = [
    TextLoader("reward_policy.txt", encoding="utf-8"),
    TextLoader("shipping_policy.txt", encoding="utf-8"),
    TextLoader("refund_policy.txt", encoding="utf-8")
]

documents = []
for loader in loaders:
    documents.extend(loader.load())

# 텍스트 분할
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
split_docs = text_splitter.split_documents(documents)


vectorstore = Chroma.from_documents(split_docs, embeddings)

# Retriever 생성
retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

# search_policy 함수 정의
def search_policy(query: str) -> str:
    """정책 문서를 검색하여 관련 내용을 반환합니다."""
    docs = retriever.invoke(query)
    if docs:
        return "\n\n".join([doc.page_content for doc in docs])
    else:
        return "관련 정책을 찾을 수 없습니다."

# 주문 데이터 (과제용 - amount 필드 추가)
orders_db = {
    "ORD001": {"status": "배송 전", "product": "노트북", "customer": "홍길동", "amount": 1500000},
    "ORD002": {"status": "배송 완료", "product": "키보드", "customer": "김철수", "amount": 80000},
    "ORD003": {"status": "배송 중", "product": "마우스", "customer": "이영희", "amount": 35000},
    "ORD004": {"status": "배송 완료", "product": "모니터", "customer": "박민수", "amount": 350000},
}

def get_order_status(order_id: str) -> str:
    """주문 상태를 조회합니다."""
    if order_id in orders_db:
        order = orders_db[order_id]
        return f"주문번호: {order_id}, 상태: {order['status']}, 상품: {order['product']}"
    return f"주문 {order_id}을 찾을 수 없습니다."

print("✅ 실습 코드 연결 완료")
print(f"   - LLM: {llm.__class__.__name__}")
print(f"   - Vector Store: {len(split_docs)}개 문서 로드")
print(f"   - Orders DB: {len(orders_db)}개 주문 데이터")
````

````python
# 환불 기록
processed_refunds = {}

# 정책에서 즉시 환불 가능한 상태 추출 (실습의 extract_max_amount와 동일한 패턴)
def extract_refund_statuses(policy_text: str) -> list:
    """LLM을 사용하여 정책 텍스트에서 즉시 환불 가능한 상태를 추출합니다."""
    prompt = f"""다음 정책 텍스트에서 '즉시 환불 가능한' 배송 상태 목록을 추출하세요.
100% 환불이 가능하거나 수령 거부 후 환불 가능한 상태만 포함합니다.
상태 이름만 쉼표로 구분하여 반환하세요. 다른 설명은 하지 마세요.

정책 텍스트:
{policy_text}

즉시 환불 가능한 상태 목록:"""

    try:
        response = llm.invoke(prompt)
        statuses = [s.strip() for s in response.content.split(",")]
        return statuses if statuses else ["배송 전", "배송 중"]
    except Exception:
        return ["배송 전", "배송 중"]  # 기본값 (추출 실패 시)

def process_refund_safe(order_id: str, reason: str) -> str:
    """
    환불 처리 (Trustworthiness 적용)

    Args:
        order_id: 주문 번호
        reason: 환불 사유

    Returns:
        처리 결과 메시지
    """
    # 1. 정책 조회 (RAG) - 제공됨
    policy = search_policy("환불 가능 조건")

    # 2. 주문 확인 - 제공됨
    if order_id not in orders_db:
        return f"❌ 환불 불가: 주문 {order_id}을 찾을 수 없습니다."

    order = orders_db[order_id]
    status = order["status"]

    # TODO: 환불 가능 조건 검증
    # 힌트: extract_refund_statuses(policy)로 즉시 환불 가능 상태 목록을 얻고,
    # status가 해당 목록에 없으면 환불 불가 메시지 반환
    # 실습의 issue_coupon_safe에서 extract_max_amount 패턴을 참고하세요.

    # ===== 답안 =====
    allowed_statuses = extract_refund_statuses(policy)
    if status not in allowed_statuses:
        return f"❌ 즉시 환불 불가: 정책 확인 결과, '{status}' 상태에서는 즉시 환불이 불가능합니다. (허용: {', '.join(allowed_statuses)})"
    # ================

    # 3. 환불 실행 - 제공됨
    processed_refunds[order_id] = {
        "amount": order["amount"],
        "reason": reason,
        "status": "완료"
    }

    return f"✅ 주문 {order_id} 환불 완료 (금액: {order['amount']:,}원, 사유: {reason})"

print("✅ 환불 처리 함수 정의 완료")
````

## Step 2: Agent 실행 + 평가 함수 호출

> 테스트 케이스를 실행하고 평가 함수를 호출하여 Agent 성과를 측정합니다.

````python
# tool 데코레이터를 사용하여 도구 정의
@tool
def search_policy_tool(query: str) -> str:
    """환불/배송 정책을 검색합니다."""
    return search_policy(query)

@tool
def get_order_status_tool(order_id: str) -> str:
    """주문 상태를 조회합니다."""
    return get_order_status(order_id)

@tool
def process_refund_tool(order_id: str, reason: str) -> str:
    """환불을 처리합니다. 주문번호와 사유를 입력합니다."""
    return process_refund_safe(order_id, reason)

# 도구 목록 (환불 도구 포함)
tools = [search_policy_tool, get_order_status_tool, process_refund_tool]

# ReAct 프롬프트 (시스템 메시지)
react_prompt = """당신은 고객 서비스 AI 에이전트입니다.
ReAct 프레임워크를 따라 요청을 처리하세요:
1. Thought → 2. Action → 3. Observation → 반복 → Final Answer"""

# Agent 생성 (LangGraph create_react_agent 사용)
agent_executor = create_react_agent(model=llm, tools=tools, prompt=react_prompt)

print("✅ Agent 생성 완료")
````

````python
# Agent 평가 함수 정의
def evaluate_agent(logs: list) -> dict:
    """Agent 성과 평가 함수
    
    평가 지표:
    - tool_order_score: 도구 호출 순서의 논리성 (0.0 ~ 1.0)
    - policy_compliance: 정책 준수 여부 (True/False)
    - task_completion: 작업 완료 여부 (True/False)
    """
    def check_logical_order(logs):
        """호출된 도구들 간의 상대적 순서가 논리적인지 확인"""
        tool_names = [log["tool"] for log in logs]
        # 논리적 우선순위: 조회(0) → 검색(1) → 실행(2)
        priority = {
            "get_order_status_tool": 0,
            "search_policy_tool": 1,
            "process_refund_tool": 2
        }
        positions = [priority[t] for t in tool_names if t in priority]
        if len(positions) <= 1:
            return 1.0  # 도구 1개 이하 → 순서 위반 불가
        # 인접 쌍이 오름차순인지 확인
        correct = sum(1 for i in range(len(positions)-1) if positions[i] <= positions[i+1])
        return correct / (len(positions) - 1)

    def check_policy_compliance(logs):
        """환불 실행 시 정책 조회를 했는지 확인"""
        tool_names = [log["tool"] for log in logs]
        has_refund = "process_refund_tool" in tool_names
        has_policy_check = "search_policy_tool" in tool_names
        if has_refund and not has_policy_check:
            return False  # 정책 확인 없이 환불 → 미준수
        return True

    def check_task_completion(logs):
        """도구 실행 결과에 성공 표시가 있는지 확인"""
        for log in logs:
            if "✅" in str(log.get("output", "")):
                return True
        return False

    return {
        "tool_order_score": round(check_logical_order(logs), 2),
        "policy_compliance": check_policy_compliance(logs),
        "task_completion": check_task_completion(logs)
    }

print("✅ 평가 함수 정의 완료")
````

아래 셀에서 Agent 실행 및 로그 추출 패턴을 확인합니다. TODO 2에서는 이 패턴을 여러 테스트 케이스에 대해 반복합니다.

````python
from langchain_core.messages import AIMessage, ToolMessage

def extract_logs_from_messages(messages: list) -> list:
    """Agent 실행 결과에서 도구 호출 로그를 추출합니다."""
    logs = []
    for i, msg in enumerate(messages):
        if isinstance(msg, AIMessage) and hasattr(msg, 'tool_calls') and msg.tool_calls:
            for tool_call in msg.tool_calls:
                tool_output = ""
                if i + 1 < len(messages) and isinstance(messages[i + 1], ToolMessage):
                    tool_output = messages[i + 1].content
                logs.append({
                    "tool": tool_call["name"],
                    "input": str(tool_call["args"]),
                    "output": tool_output
                })
    return logs

# 단일 테스트 케이스 실행 예시
case = "주문 ORD001을 단순 변심으로 환불해줘"
response = agent_executor.invoke({"messages": [{"role": "user", "content": case}]})

# 로그 추출 및 평가
logs = extract_logs_from_messages(response["messages"])
evaluation = evaluate_agent(logs)

print(f"테스트 케이스: {case}")
print(f"도구 호출 로그: {logs}")
print(f"평가 결과: {evaluation}")
````

### Concept Check: CoT(Chain-of-Thought) 패턴

CoT 패턴은 도구 호출 전에 **명시적 추론 단계**를 추가하는 방식입니다.

| 패턴 | 구조 | 특징 |
|------|------|------|
| Direct | START → agent → tools → END | 단순, 1회 실행 |
| CoT | START → reasoning → agent → tools → END | 추론 후 실행 |
| ReAct | agent ↔ tools (순환) | 반복 실행 |

#### CoT 그래프 구조
```
[START] → [reasoning] → [agent] → [tools] → [END]

※ 핵심: Direct에 reasoning 노드를 추가한 단순 선형 흐름
```

### Guided Build: 추론 노드 함수

CoT 패턴에서 사용할 `call_reasoning` 함수입니다. 이 함수는 도구 호출 전에 현재 요청을 분석하고 계획을 수립합니다.

````python
# 실습에서 정의한 AgentState, call_agent, call_tools를 재사용합니다
from langgraph.graph import StateGraph, MessagesState, START, END
from langchain_core.messages import BaseMessage, HumanMessage, AIMessage, ToolMessage, SystemMessage

# AgentState 정의 (실습과 동일 — MessagesState 상속)
class AgentState(MessagesState):
    """Agent의 상태를 정의합니다."""
    pass

# call_agent 함수 (실습과 동일)
def call_agent(state: AgentState) -> AgentState:
    response = llm.bind_tools(tools).invoke(state["messages"])
    return {"messages": [response]}

# call_tools 함수 (실습과 동일)
def call_tools(state: AgentState) -> AgentState:
    last_message = state["messages"][-1]
    tool_messages = []
    for tool_call in last_message.tool_calls:
        for t in tools:
            if t.name == tool_call["name"]:
                result = t.invoke(tool_call["args"])
                tool_messages.append(ToolMessage(
                    content=str(result),
                    tool_call_id=tool_call["id"]
                ))
                break
    return {"messages": tool_messages}

# Guided Build: 추론 노드 함수 (제공 코드)
def call_reasoning(state: AgentState) -> AgentState:
    """CoT: 도구 호출 전 추론 단계"""
    reasoning_prompt = """현재 요청을 분석하고, 어떤 도구를 어떤 순서로 사용해야 하는지 단계별로 설명하세요.
    
    사용 가능한 도구:
    - search_policy_tool: 환불/배송 정책 검색
    - get_order_status_tool: 주문 상태 조회
    - process_refund_tool: 환불 처리
    """
    messages = state["messages"] + [HumanMessage(content=reasoning_prompt)]
    response = llm.invoke(messages)
    return {"messages": [response]}

# 추론 함수 테스트
test_state = {"messages": [HumanMessage(content="주문 ORD001을 환불해줘")]}
result = call_reasoning(test_state)
print("추론 결과:", result["messages"][-1].content)
````

  4. 컴파일: `workflow.compile()`

````python
# TODO: CoT 패턴 StateGraph 구현
# 요구사항: 실습에서 배운 StateGraph 구성 방식을 활용하여 CoT 그래프를 완성하세요.
# 
# 힌트: 
# 1. StateGraph(AgentState) 생성
# 2. 노드 추가: add_node("reasoning", call_reasoning), add_node("agent", call_agent), add_node("tools", call_tools)
# 3. 엣지 연결: add_edge(START, "reasoning"), reasoning → agent, agent → tools, tools → END
# 4. 컴파일: workflow.compile()

from langgraph.graph import StateGraph, START, END

workflow_cot = StateGraph(AgentState)

# ===== 답안 =====
# 노드 추가
workflow_cot.add_node("reasoning", call_reasoning)
workflow_cot.add_node("agent", call_agent)
workflow_cot.add_node("tools", call_tools)

# ★ CoT 핵심: 단순한 선형 흐름 (reasoning 추가)
workflow_cot.add_edge(START, "reasoning")
workflow_cot.add_edge("reasoning", "agent")
workflow_cot.add_edge("agent", "tools")
workflow_cot.add_edge("tools", END)
# ================

agent_cot = workflow_cot.compile()

print("✅ CoT 패턴 Agent 생성 완료")
````

3. 도구 실행 후 종료

````python
# CoT 패턴 그래프 시각화
print("=== CoT 패턴 그래프 구조 ===")
print(agent_cot.get_graph().draw_mermaid())

# CoT 패턴 테스트
print("\n=== CoT 패턴 테스트 ===")
test_request = "주문 ORD001을 단순 변심으로 환불해줘."
initial_state = {"messages": [HumanMessage(content=test_request)]}
result = agent_cot.invoke(initial_state)

print("\n메시지 흐름:")
for i, msg in enumerate(result["messages"]):
    msg_type = type(msg).__name__
    content_preview = msg.content[:100] if msg.content else "(도구 호출)"
    print(f"  [{i+1}] {msg_type}: {content_preview}...")

print("\n최종 응답:")
print(result["messages"][-1].content)
````

### CoT 패턴 비교 평가

Step 2에서 `agent_executor`(ReAct)로 실행한 동일한 테스트 케이스를 `agent_cot`(CoT)로 다시 실행하여 평가 결과를 비교합니다.

**비교 포인트:**
- `tool_order_score`: CoT의 reasoning 단계가 도구 호출 순서를 개선하는지 확인
- `task_completion`: 추론 단계 추가가 작업 완료율에 영향을 주는지 확인

````python
# CoT 패턴으로 동일 테스트 케이스 실행 및 비교
results_cot = []

for i, case in enumerate(test_cases, 1):
    print(f"\n{'='*50}")
    print(f"[CoT] 테스트 케이스 {i}: {case}")
    print('='*50)

    initial_state = {"messages": [HumanMessage(content=case)]}
    response = agent_cot.invoke(initial_state)
    output = response["messages"][-1].content

    logs = extract_logs_from_messages(response["messages"])
    evaluation = evaluate_agent(logs)

    results_cot.append({
        "case": case,
        "output": output,
        "evaluation": evaluation
    })

    print(f"평가 결과: {evaluation}")

# 비교 출력
print(f"\n{'='*60}")
print("📊 ReAct vs CoT 평가 결과 비교")
print('='*60)
for i, (r, c) in enumerate(zip(results, results_cot), 1):
    print(f"\n[케이스 {i}] {r['case'][:30]}...")
    print(f"  {'지표':<20} {'ReAct':<15} {'CoT':<15}")
    print(f"  {'-'*50}")
    for key in ['tool_order_score', 'policy_compliance', 'task_completion']:
        rv = r['evaluation'][key]
        cv = c['evaluation'][key]
        marker = " ✦" if rv != cv else ""
        print(f"  {key:<20} {str(rv):<15} {str(cv):<15}{marker}")
````

### 비교 결과 해석

위 비교표에서 ✦ 표시는 ReAct와 CoT 간 차이가 발생한 지표입니다.

#### 주요 관찰 포인트

| 관찰 항목 | 해석 |
|-----------|------|
| `tool_order_score`가 CoT에서 높아졌다면 | reasoning 단계에서 도구 사용 계획을 세운 효과 — 논리적 순서로 도구를 호출 |
| `tool_order_score`가 동일하다면 | ReAct의 반복 추론으로도 충분히 올바른 순서를 찾음 |
| `task_completion`이 달라졌다면 | CoT의 단일 패스 구조(순환 없음) 때문에 복합 요청 처리에 차이 발생 가능 |

#### CoT vs ReAct 트레이드오프

| 측면 | CoT | ReAct |
|------|-----|-------|
| **도구 순서** | 계획 후 실행 → 순서 정확도 높음 | 시행착오 가능 → 순서 불안정 |
| **복합 요청** | 단일 패스 → 처리 한계 | 순환 구조 → 반복 처리 가능 |
| **토큰 비용** | reasoning 단계 추가 비용 | 반복 시 누적 비용 |

> **핵심**: 단순 요청에는 CoT가 효율적이고, 복합 요청에는 ReAct가 유리합니다.
