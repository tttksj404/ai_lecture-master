## Step 1: Python Class 기초

### Concept Check: 클래스와 객체 지향 프로그래밍

**클래스(Class)란?**

클래스는 객체를 만들기 위한 **설계도(blueprint)** 입니다. 마치 붕어빵 틀처럼, 클래스라는 틀로 여러 개의 객체(붕어빵)를 만들 수 있습니다.

**핵심 용어 정리**

1. **클래스(Class)**: 객체의 구조와 동작을 정의하는 설계도
   - 예: 자동차 설계도, 붕어빵 틀

2. **인스턴스(Instance)**: 클래스로 만든 실제 객체
   - 예: 설계도로 만든 실제 자동차, 틀로 만든 실제 붕어빵

3. **속성(Attribute)**: 객체가 가지는 데이터(상태)
   - 예: 자동차의 색상, 모델, 속도

4. **메서드(Method)**: 객체가 수행하는 동작(함수)
   - 예: 자동차의 accelerate(가속), brake(브레이크)

**왜 클래스를 사용하나요?**

- **코드 재사용**: 같은 구조의 객체를 여러 개 만들 때 코드 중복을 줄입니다.
- **구조화**: 관련된 데이터와 함수를 하나로 묶어 관리합니다.
- **유지보수**: 설계도(클래스)만 수정하면 모든 객체에 반영됩니다.

**기본 문법**

```python
class ClassName:
    def __init__(self, param1, param2):  # 생성자: 객체 생성 시 호출
        self.attribute1 = param1  # 속성 정의
        self.attribute2 = param2
    
    def method_name(self):  # 메서드 정의
        # 동작 구현
        return something
```

**self란?**
- `self`는 인스턴스 자기 자신을 가리키는 참조입니다.
- 메서드 내에서 인스턴스의 속성에 접근할 때 사용합니다.
- 메서드 정의 시 첫 번째 매개변수로 반드시 작성하지만, 호출 시에는 생략합니다.

### Guided Build: 간단한 클래스 예제

아래 예제는 `Person` 클래스를 정의하고 인스턴스를 생성하는 과정을 보여줍니다.

````python
# Person 클래스 정의
class Person:
    def __init__(self, name, age):
        self.name = name  # 속성: 이름
        self.age = age    # 속성: 나이

    def introduce(self):  # 메서드: 자기소개
        return f"안녕하세요, 저는 {self.name}이고 {self.age}살입니다."

# 인스턴스 생성
person1 = Person("철수", 25)
person2 = Person("영희", 30)

# 메서드 호출
print(person1.introduce())
print(person2.introduce())

# 속성 접근
print(f"{person1.name}의 나이: {person1.age}")
````

### Guided Build: Calculator 클래스 예제

아래는 간단한 계산기 클래스입니다. 메서드가 어떻게 작동하는지 확인하세요.

````python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

# 테스트
calc = Calculator()
print(f"10 + 5 = {calc.add(10, 5)}")
print(f"10 - 5 = {calc.subtract(10, 5)}")
print(f"10 * 5 = {calc.multiply(10, 5)}")
````

  - `self.history.append()`로 거래 내역 추가
  - 타입 힌팅 사용 권장: `def __init__(self, owner: str, balance: int = 0)`

````python
class BankAccount:
    """은행 계좌 클래스
    Attributes:
        owner (str): 계좌 소유자 이름
        balance (int): 계좌 잔액
        history (list): 거래 내역 리스트
    """
    def __init__(self, owner: str, balance: int = 0):
        # 1. 계좌소유자, 계좌잔액, 거래내역 리스트를 `owner`, `balance`, `history` attribute으로 저장해주세요
        self.owner = owner
        self.balance = balance
        self.history = []


    def deposit(self, amount: int):
        # 2. 입금 처리 (0 이하면 "입금 실패", 성공하면 "입금 완료" 출력 및 거래 내역 추가)
        if amount <=0:
            print("입금 실패")
            return
        self.balance += amount
        self.history.append(f"입금: {amount}원")
        print("입금 완료")

    def withdraw(self, amount: int):
        # 3. 출금 처리 (0 이하면 "출금 실패", 잔액 부족하면 "잔액 부족", 성공하면 "출금 완료" 출력 및 거래 내역 추가)
        if amount<=0:
            print("출금 실패")
            return
        if amount>self.balance:
            print("잔액 부족")
        self.balance-=amount
        self.history.append(f"출금: {amount}원")
        print("출금 완료")

    def show_balance(self):
        # 4. "현재 잔액: {balance}원" 형식으로 출력
        print(f"현재 잔액: {self.balance}원")

    def get_transaction_history(self) -> list:
        # 5. 거래 내역이 없으면 "거래 내역이 없습니다." 출력, 있으면 모든 내역 출력 후 리스트 반환
        if not self.history:
            print("거래 내역이 없습니다.")
        else:
            print(*self.history, sep="\n")
        return self.history

    def __str__(self):
        # 6. "[계좌 소유자: {owner}, 잔액: {balance}원]" 형식으로 반환
        return f"[계좌 소유자: {self.owner}, 잔액: {self.balance}원]"

# 간단한 테스트
account = BankAccount("김철수", 10000)
account.deposit(5000)
account.withdraw(3000)
account.show_balance()
print(account)
````

### Concept Check: Pandas DataFrame이란?

**Pandas란?**

Pandas는 Python에서 데이터 분석을 위해 가장 많이 사용되는 라이브러리입니다. 표 형태의 데이터를 다루는 데 특화되어 있습니다.

**DataFrame이란?**

DataFrame은 Pandas의 핵심 자료구조로, 엑셀의 시트나 SQL 테이블처럼 **행(row)과 열(column)로 구성된 2차원 테이블**입니다.

```
   total_bill   tip     sex smoker  day    time  size
0       16.99  1.01  Female     No  Sun  Dinner     2
1       10.34  1.66    Male     No  Sun  Dinner     3
2       21.01  3.50    Male     No  Sun  Dinner     3
```

**기본 용어**
- **행(Row)**: 각각의 데이터 레코드 (위 예시에서 0, 1, 2...)
- **열(Column)**: 각각의 속성/변수 (total_bill, tip, sex...)
- **인덱스(Index)**: 각 행을 식별하는 번호 또는 라벨

**데이터 확인 함수**

1. **df.head(n)**: 처음 n개 행 보기 (기본값 5)
2. **df.info()**: 데이터 타입, 결측치, 메모리 정보 확인
3. **df.describe()**: 수치형 컬럼의 통계 요약(평균, 표준편차, 최소/최대값 등)
4. **df.shape**: (행 개수, 열 개수) 튜플로 반환
5. **df.columns**: 컬럼 이름 리스트

**데이터 선택 방법**

1. **컬럼 선택**: `df['컬럼명']` 또는 `df[['컬럼1', '컬럼2']]`
2. **loc (라벨 기반)**: `df.loc[행_라벨, 열_라벨]`
3. **iloc (위치 기반)**: `df.iloc[행_번호, 열_번호]`

**조건 필터링 (불리언 인덱싱)**

조건식을 사용하여 특정 조건을 만족하는 행만 선택할 수 있습니다.

```python
# 팁이 5달러 이상인 행만 선택
df[df['tip'] >= 5]

# 성별이 Male인 행만 선택
df[df['sex'] == 'Male']

# 여러 조건 조합 (AND: &, OR: |)
df[(df['tip'] >= 5) & (df['sex'] == 'Male')]
```

**기초 집계 함수**

- **mean()**: 평균
- **sum()**: 합계
- **count()**: 개수
- **min() / max()**: 최소/최대값
- **std()**: 표준편차

```python
df['tip'].mean()  # 팁 평균
df['size'].sum()  # 인원수 합계
```

## Step 4: Pandas 심화 (groupby, pivot, melt)

### Concept Check: groupby, pivot, melt

- **groupby**: 그룹별 집계 및 분석
- **pivot**: wide-format 테이블 생성
- **melt**: wide-format → long-format 변환
- **apply**: 사용자 정의 함수 적용
- 복합 조건 필터링 및 데이터 해석

이번 Step에서는 Pandas의 심화 기능을 활용하여 데이터를 변환하고 복잡한 분석을 수행합니다.

**groupby: 그룹별 집계**

`groupby`는 데이터를 특정 기준(컬럼)에 따라 그룹으로 나누고, 각 그룹에 대해 집계 함수를 적용합니다.

**기본 문법**
```python
df.groupby('컬럼명')['집계_컬럼'].집계함수()
```

**예시**
```python
# 성별(sex)별 팁 평균
df.groupby('sex')['tip'].mean()

# 요일(day)별 총 금액 합계
df.groupby('day')['total_bill'].sum()

# 여러 컬럼 기준 그룹화
df.groupby(['sex', 'day'])['tip'].mean()
```

**pivot: wide-format 테이블 생성**

`pivot`은 특정 컬럼의 값을 행과 열로 재배치하여 크로스탭(교차표) 형태로 만듭니다.

**기본 문법**
```python
df.pivot(index='행_컬럼', columns='열_컬럼', values='값_컬럼')
```

**예시**
```python
# 성별(행) × 요일(열) 팁 평균 테이블
df.groupby(['sex', 'day'])['tip'].mean().reset_index().pivot(
    index='sex', columns='day', values='tip'
)
```

**melt: wide → long 변환**

`melt`는 여러 컬럼에 분산된 데이터를 하나의 컬럼으로 모아서 "길게" 늘립니다.

**기본 문법**
```python
pd.melt(df, id_vars=['고정_컬럼'], value_vars=['변환_컬럼들'],
        var_name='변수명', value_name='값명')
```

**예시**
```python
# total_bill과 tip을 하나의 컬럼으로 변환
pd.melt(df, id_vars=['size'], value_vars=['total_bill', 'tip'],
        var_name='item', value_name='amount')
```

**언제 사용하나요?**
- groupby: 카테고리별 통계를 빠르게 계산할 때
- pivot: 데이터를 표 형태로 보기 좋게 정리할 때
- melt: 시각화나 특정 분석을 위해 데이터 형태를 변환할 때
